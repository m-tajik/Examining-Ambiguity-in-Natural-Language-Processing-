```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Define the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to convert NLTK's POS tags to WordNet's POS tags
def get_wordnet_pos(nltk_pos):
    if nltk_pos.startswith('J'):
        return wordnet.ADJ
    elif nltk_pos.startswith('V'):
        return wordnet.VERB
    elif nltk_pos.startswith('N'):
        return wordnet.NOUN
    elif nltk_pos.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # by default, assume it's a noun

# Load the dataset
import pandas as pd


# Lemmatize each sentence in the dataset
lemmatized_sentences = []
for sentence in dataframe['context_list_without_html']:
    # Tokenize the sentence and get the POS tags
    nltk_tokens = nltk.word_tokenize(sentence)
    nltk_pos = nltk.pos_tag(nltk_tokens)

    # Convert the NLTK POS tags to WordNet POS tags
    wordnet_pos = [(token, get_wordnet_pos(pos)) for token, pos in nltk_pos]

    # Lemmatize each word in the sentence
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos) for token, pos in wordnet_pos]

    # Join the lemmatized tokens back into a sentence
    lemmatized_sentence = ' '.join(lemmatized_tokens)
    lemmatized_sentences.append(lemmatized_sentence)

# Add the lemmatized sentences to the dataset
dataframe['lemmatized_text'] = lemmatized_sentences


```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
if Version(sklearn_version) < '0.18':
    from sklearn.grid_search import GridSearchCV
else:
    from sklearn.model_selection import GridSearchCV

tfidf = TfidfVectorizer(strip_accents=None,
                        lowercase=False,
                        preprocessor=None)

param_grid = [{'vect__ngram_range': [(1, 1)],
              #  'vect__stop_words': [stop, None],
               'vect__tokenizer': [word_tokenize],
               'clf__penalty': ['l1', 'l2'],
               'clf__C': [1.0, 10.0, 100.0]},
              {'vect__ngram_range': [(1, 1)],
              #  'vect__stop_words': [stop, None],
               'vect__tokenizer': [word_tokenize],
               'vect__use_idf':[False],
               'vect__norm':[None],
               'clf__penalty': ['l1', 'l2'],
               'clf__C': [1.0, 10.0, 100.0]},
              ]

lr_tfidf = Pipeline([('vect', tfidf),
                     ('clf', LogisticRegression(random_state=0))])

gs_lr_tfidf = GridSearchCV(lr_tfidf, param_grid,
                           scoring='accuracy',
                           cv=5,
                           verbose=1,
                           n_jobs=-1)
```

```python
def iqv(x: list):
  score_array = []
  for i in x:
      k = len(i)
      score = k*(1-np.sum(np.square(np.array(i))))/(k-1)
      score_array.append(score)
  return(score_array)
```

```python
import pickle,os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
unique_words = set(df2['word'])
scores_list=[]
u_words=[]
accu_list=[]
df3=pd.DataFrame()
uniquedf=pd.DataFrame()
for word in unique_words:
    # Create a subset of the data containing only the rows with the current word
    word_data = df2[df2['word'] == word]

    # Split the data into training and testing sets
    Xtrain, Xtest, ytrain, ytest = train_test_split(word_data['train'], word_data['sense_id'], test_size=0.2)

    # Create a TF-IDF vectorizer
    tfidf = TfidfVectorizer()

    # Vectorize the training and testing data
    vectorizer=tfidf.fit(Xtrain)
    Xtrain_tfidf = vectorizer.transform(Xtrain)
    Xtest_tfidf = vectorizer.transform(Xtest)
    len(ytrain)

    # Train a logistic regression model
    lr = LogisticRegression()
    modelobject = lr.fit(Xtrain_tfidf, ytrain)

    # Evaluate the model
    accuracy = lr.score(Xtest_tfidf, ytest)

    ypred = modelobject.predict(Xtest_tfidf)
    udf=pd.DataFrame()
    u_words.append(word)
    accu_list.append(accuracy)

    uniquedf = pd.concat([uniquedf, udf], ignore_index=True)
    print(f"Accuracy for {word}: {accuracy}")

    probs=modelobject.predict_proba(Xtest_tfidf)
    Xtestdf=pd.DataFrame()
    Xtestdf['iqv']=iqv(probs)
    Xtestdf['train']=np.array(Xtest)

    df3 = pd.concat([df3, Xtestdf], ignore_index=True)
    modelobject_filename = os.path.join("/content/drive/MyDrive/ModelObjects",str(word+".pkl"))
    vectorizer_filename = os.path.join("/content/drive/MyDrive/Colab Notebooks/Vectorizers",str(word+".pkl"))
    pickle.dump(modelobject, open(modelobject_filename, 'wb'))
    pickle.dump(vectorizer, open(vectorizer_filename, 'wb'))


    # Assume that `y_true` and `y_pred` are the true and predicted labels, respectively
    cm = confusion_matrix(ytest, ypred)
    # Create a heatmap using seaborn
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')

    # Add labels and title
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix for '+str(word))

    # Show the plot
    plt.show()
```
